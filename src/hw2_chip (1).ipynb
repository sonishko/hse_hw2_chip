#Выравнивание на хромосому


```python
!wget https://hgdownload.soe.ucsc.edu/goldenPath/hg19/chromosomes/chr15.fa.gz
!gzip -d chr15.fa.gz
```

```python
%%time
!bowtie2-build chr15.fa chromosome_index
```

```python
!mkdir bowtie2_res
```

```python
%%time
!bowtie2  -p 2 \
          -x chromosome_index \
          -U ENCFF000VIP.fastq \
          -S bowtie2_res/res_VIP.sam
```

```python
%%time
!bowtie2  -p 2 \
          -x chromosome_index \
          -U ENCFF000VIR.fastq \
          -S bowtie2_res/res_VIR.sam
```

```python
%%time
!bowtie2  -p 2 \
          -x chromosome_index \
          -U ENCFF000VJK.fastq \
          -S bowtie2_res/res_VJK.sam
```

#Peak calling

```python
!mkdir macs2
```

```python
!macs2 callpeak --broad -t bowtie2_res/res_VIP.sam \
    -c bowtie2_res/res_VJK.sam \
 	  -f SAM \
	  --outdir macs2_VIP
```

```python
!macs2 callpeak --broad -t bowtie2_res/res_VIR.sam \
    -c bowtie2_res/res_VJK.sam \
 	  -f SAM \
	  --outdir macs2_VIR
```
